```python
# Install and load pandas and matplotlib
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load your CSV files #

# Create a dataframe named 'daily_activity' and read in one of the CSV files from the dataset. 
# Remember, you can name your dataframe something different, and you can also save your CSV file under a different name as well.
daily_activity = pd.read_csv("dailyActivity_merged.csv")

# Create another dataframe for the sleep data.
sleep_day = pd.read_csv("sleepDay_merged.csv")
```

```python

# Explore a few key tables #

# Take a look at the daily_activity data.
print(daily_activity.head())

# Identify all the columns in the daily_activity data.
print(daily_activity.columns)

# Take a look at the sleep_day data.
print(sleep_day.head())

# Identify all the columns in the daily_activity data.
print(sleep_day.columns)
```

```python
# Understanding some summary statistics #

# How many unique participants are there in each dataframe?
# It looks like there may be more participants in the daily activity
# dataset than the sleep dataset.
print(daily_activity['Id'].nunique())
print(sleep_day['Id'].nunique())

# How many observations are there in each dataframe?
print(daily_activity.shape[0])
print(sleep_day.shape[0])

# What are some quick summary statistics we'd want to know about each data frame?
# For the daily activity dataframe:
print(daily_activity[['TotalSteps', 'TotalDistance', 'SedentaryMinutes']].describe())

# For the sleep dataframe:
print(sleep_day[['TotalSleepRecords', 'TotalMinutesAsleep', 'TotalTimeInBed']].describe())
```

```python
# What does this tell us about how this sample of people's activities?

# Plotting a few explorations #

# What's the relationship between steps taken in a day and sedentary minutes?
# How could this help inform the customer segments that we can market to?
# E.g. position this more as a way to get started in walking more?
# Or to measure steps that you're already taking?
plt.scatter(daily_activity['TotalSteps'], daily_activity['SedentaryMinutes'])
plt.xlabel('Total Steps')
plt.ylabel('Sedentary Minutes')
plt.show()
```

```python

# What's the relationship between minutes asleep and time in bed?
# You might expect it to be almost completely linear - are there any unexpected trends?
plt.scatter(sleep_day['TotalMinutesAsleep'], sleep_day['TotalTimeInBed'])
plt.xlabel('Total Minutes Asleep')
plt.ylabel('Total Time In Bed')
plt.show()
```

```python
# Merging these two datasets together #

combined_data = pd.merge(sleep_day, daily_activity, on="Id")

# Take a look at how many participants are in this data set.
print(combined_data['Id'].nunique())
```

Note that there were more participant Ids in the daily activity dataset that have been filtered out using merge. Consider using 'outer_join' to keep those in the dataset.
